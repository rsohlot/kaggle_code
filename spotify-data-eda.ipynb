```python
track_count = data1['track_uri'].value_counts().to_frame().reset_index()
track_count.rename(columns = {'track_uri' : 'count', 'index' : 'track_uri'},inplace = True)
track_count.sort_values(['count'],ascending=False,inplace=True)
track_count['count_as_str'] = track_count['count'].astype(str)
print('track_count: ',track_count.shape)
track_count.head(2)


# track_occurance_count = track_count['count']
```
